```python
from teili.core.groups import Neurons, Connections
from brian2 import PoissonGroup, SpikeGeneratorGroup, Hz, ms
import numpy as np


def quantize_weights_number(synapse_group, n_weights=20, n_hw_neurons=1, spikegen_groups=None, debug=False):
    connectivity_matrix =  np.zeros((100, 49))
    if np.asarray(synapse_group.w_plast).all() != 1:
        weights = synapse_group.w_plast
    else:
        weights = synapse_group.weight

    if synapse_group.source.name in spikegen_groups:
        if (synapse_group.weight >= 0).any():
            syn = c.modules.CtxDynapse.DynapseCamType.FAST_EXC
        else:
            syn = c.modules.CtxDynapse.DynapseCamType.FAST_INH
        weights = np.reshape(
            weights, (synapse_group.source.N, synapse_group.target.N))
        if n_weights > synapse_group.source.N:
            n_weights = synapse_group.source.N

        source_list = []
        target_list = []
        syn_type = []
        for offset in range(n_hw_neurons):
            for target in range(np.shape(weights)[1]):
                sorted_indices = np.argsort(weights[:, target])
                for source in sorted_indices[-n_weights:]:
                    if weights[source, target] == 0:
                        continue
                    else:
                        source_list.append(np.int(source))
                        target_list.append(
                            np.int(target * n_hw_neurons + offset))

                        syn_type.append(syn)
                        if debug:
                            print('Connecting {} to {} using {}'.format(np.int(source),
                                                                       np.int(target * n_hw_neurons + offset),
                                                                       syn.name))
                            connectivity_matrix[source, target] = 1
        if debug:
            np.save('connectivity_matrx.npy', connectivity_matrix)
        return source_list, target_list, syn_type

    if (synapse_group.weight >= 0).any():
        syn = c.modules.CtxDynapse.DynapseCamType.SLOW_EXC
    else:
        syn = c.modules.CtxDynapse.DynapseCamType.FAST_INH

    if len(weights) < synapse_group.source.N * synapse_group.target.N:
        source_list = []
        target_list = []
        syn_type = []
        
        for offset in range(n_hw_neurons):
            for ii in range(len(synapse_group.j)):
                source = synapse_group.i[ii]
                target = synapse_group.j[ii]
                source_list.append(
                    np.int(source * n_hw_neurons) + offset)
                target_list.append(
                    np.int(target * n_hw_neurons) + offset)
                syn_type.append(syn)
                if debug:
                    print('Connecting {} to {} using {}'.format(np.int(source * n_hw_neurons) + offset,
                                                                    np.int(target * n_hw_neurons) + offset,
                                                                    syn.name))
        
    else:
        weights = np.reshape(
            weights, (synapse_group.source.N, synapse_group.target.N))
        if n_weights > synapse_group.source.N:
            n_weights = synapse_group.source.N

        source_list = []
        target_list = []
        syn_type = []
        for offset in range(n_hw_neurons):
            for target in range(np.shape(weights)[1]):
                sorted_indices = np.argsort(weights[:, target])
                for source in sorted_indices[-n_weights:]:
                    if weights[source, target] == 0:
                        continue
                    else:
                        source_list.append(
                            np.int(source * n_hw_neurons) + offset)
                        target_list.append(
                            np.int(target * n_hw_neurons) + offset)
                        syn_type.append(syn)
                        if debug:
                            print('Connecting {} to {} using {}'.format(np.int(source * n_hw_neurons) + offset,
                                                                        np.int(target * n_hw_neurons) + offset,
                                                                        syn.name))
    return source_list, target_list, syn_type


def convert(Network, start_neurons, n_hw_neurons=1, isi_base=90, quantize_weights=True, debug=False):
    """Recreates detected teili objects of the given network on
    DYNAP-SE using CTXCTL interface

    Arguments:
        Netork(teiliNetwork)
        start_neurons(list)

    """
    neuron_groups = {att.name: att for att in Network.__dict__['objects']
                     if type(att) == Neurons}
    conn_groups = {att.name: att for att in Network.__dict__['objects']
                   if type(att) == Connections}
    poisson_groups = {att.name: att for att in Network.__dict__['objects']
                      if type(att) == PoissonGroup}
    spikegen_groups = {att.name: att for att in Network.__dict__['objects']
                       if type(att) == SpikeGeneratorGroup}

    assert len(neuron_groups) + len(poisson_groups) + len(spikegen_groups) == len(start_neurons),\
        "Number of populations must match the list of indices"
        
    print('Clearing cams...')
    c.modules.CtxDynapse.dynapse.clear_cam(0)
    c.modules.CtxDynapse.dynapse.clear_cam(1)
    c.modules.CtxDynapse.dynapse.clear_cam(2)
    c.modules.CtxDynapse.dynapse.clear_cam(3)
#     c.modules.CtxDynapse.model.apply_diff_state()
    print('Cams cleared!')

    model = c.modules.CtxDynapse.model
    model.apply_diff_state()
    neurons = model.get_shadow_state_neurons()
    virtual_model = c.modules.CtxDynapse.VirtualModel()
    virtual_neurons = virtual_model.get_neurons()

    dynapse_groups = {}

    if debug is True:
        print('Allocating on-chip populations (%d neuron groups detected)'
              % (len(neuron_groups)))

    for n_group in neuron_groups:
        dynapse_groups.update({n_group:
                               [neurons[i] for i in range(len(neurons)) if
                                i >= start_neurons[n_group] and
                                i < start_neurons[n_group] +
                                neuron_groups[n_group].N * n_hw_neurons]})

    if poisson_groups:
        for p_group in poisson_groups:
            dynapse_groups.update({p_group:
                                   [virtual_neurons[i] for i in range(len(virtual_neurons)) if
                                    i >= start_neurons[p_group] and
                                    i < start_neurons[p_group] +
                                    poisson_groups[p_group].N]})

    for s_group in spikegen_groups:
        dynapse_groups.update({s_group:
                               [virtual_neurons[i] for i in range(len(virtual_neurons)) if
                                i >= start_neurons[s_group] and
                                i < start_neurons[s_group] +
                                spikegen_groups[s_group].N]})

    if debug is True:
        print('Connecting on-chip populations (%d connection groups detected)'
              % (len(conn_groups)))

    for conn in conn_groups:
        if debug is True:
                print('Connecting %s to %s' % (conn_groups[conn].source.name,
                                               conn_groups[conn].target.name))
                
        if conn_groups[conn].source.name in dynapse_groups and \
                conn_groups[conn].target.name in dynapse_groups:

            if quantize_weights:
                source_list, target_list, syn_type = quantize_weights_number(conn_groups[conn],
                                                                             n_weights=20,
                                                                             n_hw_neurons=n_hw_neurons,
                                                                             spikegen_groups=spikegen_groups,
                                                                             debug=debug)
                connect_populations(dynapse_groups[conn_groups[conn].source.name],
                                    dynapse_groups[
                                        conn_groups[conn].target.name],
                                    source_list,
                                    target_list,
                                    syn_type)
            else:
                syn_type = []
                for w in conn_groups[conn].weight:
                    if w >= 0:
                        syn_type.append(SynapseTypes.SLOW_EXC)
                    else:
                        syn_type.append(SynapseTypes.SLOW_INH)

                connect_populations(dynapse_groups[conn_groups[conn].source.name],
                                    dynapse_groups[
                                        conn_groups[conn].target.name],
                                    conn_groups[conn].i,
                                    conn_groups[conn].j,
                                    syn_type)

            
    # Set input
    set_spikes_fpga(spikegen_groups, start_neurons, isi_base=isi_base)

    return dynapse_groups


def connect_populations(pre_pop, post_pop, i, j, syn_type):
    """
    Wrapper function for connecting populations given lists of indices
    """
    assert len(i) == len(j), "index lists i and j must be of equal length"
    connector = c.modules.NeuronNeuronConnector.DynapseConnector()
    for index in range(len(i)):
        connector.add_connection(pre=pre_pop[i[index]], post=post_pop[
                                 j[index]], synapse_type=syn_type[index])


def set_spikes_fpga(spikegen_groups, start_neurons, isi_base=90, poisson=False):
    print('Setting input')
    model = c.modules.CtxDynapse.model
    fpgaModules = model.get_fpga_modules()
    if poisson:
        spikegen = fpgaModules[0]
        for p_group in spikegen_groups:
            for i in range(spikegen_groups[p_group].N):
                spikegen.write_poisson_rate_hz(i, spikegen_groups[p_group].rates[
                                               i - start_neurons[p_group]] / Hz)
    else:
        max_isi_us = (2**16 - 1) * isi_base / 90
        max_fpga_len = 2**15 - 1
        for s_group in spikegen_groups:
            spikegen = fpgaModules[1]
            fpga_nrn_ids = spikegen_groups[s_group]._neuron_index[1:]

            spiketimes = (
                spikegen_groups[s_group]._spike_time * 10**(6)).astype(int)
            fpga_isi = np.diff(spiketimes)
            fpga_events = []
            assert all(fpga_isi <
                       max_isi_us), 'ISI larger than encodable isi with isi_base={}.\
                             Add dummy neuron'.format(isi_base)

            for idx_isi, isi in enumerate(fpga_isi):
                event = c.modules.CtxDynapse.FpgaSpikeEvent()
                event.target_chip = 0
                event.core_mask = 1
                event.neuron_id = fpga_nrn_ids[idx_isi]
                event.isi = isi
                fpga_events.append(event)
                if idx_isi % 5000 == 0:
                    print('Loading events...')
        print('Events loaded.')
        print('Setting variable ISI...')
        spikegen.set_variable_isi(True)
        spikegen.set_isi_multiplier(isi_base)
        print('Events loaded into FpgaSpikeEvent list.')
        spikegen.preload_stimulus(fpga_events)
        print('Setting repeat mode...')
        spikegen.set_repeat_mode(1)
        print('About to start stimulation...')
        spikegen.start()
```
